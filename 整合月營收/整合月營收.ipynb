```python
import os
import pandas as pd

# === è³‡æ–™å¤¾è¨­å®š ===
INPUT_DIR = os.getcwd()
OUTPUT_DIR = os.path.join(os.getcwd(), "mops_merged")
os.makedirs(OUTPUT_DIR, exist_ok=True)

# === æ‰¾å‡ºæ‰€æœ‰ mops_YYYY_MM.csv ===
csv_files = [f for f in os.listdir(INPUT_DIR) if f.startswith("mops_") and f.endswith(".csv")]

if not csv_files:
    print("âŒ æ‰¾ä¸åˆ°ä»»ä½• mops_YYYY_MM.csv æª”æ¡ˆ")
    exit()

all_data = []

print(f"ğŸ“¥ é–‹å§‹è®€å– {len(csv_files)} å€‹æª”æ¡ˆ...")
for file in sorted(csv_files):
    file_path = os.path.join(INPUT_DIR, file)
    try:
        df = pd.read_csv(file_path, dtype=str)
        all_data.append(df)
        print(f"âœ… è®€å–ï¼š{file}")
    except Exception as e:
        print(f"âš ï¸ éŒ¯èª¤è®€å– {file}ï¼š{e}")

# === åˆä½µæ‰€æœ‰è³‡æ–™ ===
merged_df = pd.concat(all_data, ignore_index=True)

# === åŠ ä¸€å€‹æ•¸å€¼æ¬„ä½ç”¨æ–¼æ¯”è¼ƒï¼ˆä¸æ”¹åŸæ¬„ä½å…§å®¹ï¼‰===
merged_df["_ç‡Ÿæ”¶æ•¸å€¼"] = (
    merged_df["ç‡Ÿæ¥­æ”¶å…¥-ç•¶æœˆç‡Ÿæ”¶"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .replace("", pd.NA)
    .astype(float)
)

# === æ ¹æ“šã€Œå¹´åº¦ + æœˆä»½ + å…¬å¸ä»£è™Ÿã€åˆ†çµ„ï¼Œå–æœ€å¤§å–®æœˆç‡Ÿæ”¶é‚£ç­† ===
idx = (
    merged_df
    .groupby(["å¹´åº¦", "æœˆä»½", "å…¬å¸ä»£è™Ÿ"])["_ç‡Ÿæ”¶æ•¸å€¼"]
    .idxmax()
)

max_df = merged_df.loc[idx].drop(columns=["_ç‡Ÿæ”¶æ•¸å€¼"]).reset_index(drop=True)

# === å„²å­˜è¼¸å‡º ===
output_file = os.path.join(OUTPUT_DIR, "all_mops_max_per_month.csv")
max_df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f"\nğŸ‰ å®Œæˆï¼æ¯æœˆæ¯å…¬å¸ä¿ç•™æœ€å¤§ç‡Ÿæ”¶è³‡æ–™ï¼Œå…± {len(max_df)} ç­†")
print(f"ğŸ“¦ æª”æ¡ˆå„²å­˜æ–¼ï¼š{output_file}")
```
